```python
#le a tabela substituindo todos os 0.0 por nan
tb2=ascii.read(mydir+exfile,fill_values=[('0.0',str(np.nan))])
```

```python
t=np.array(tb2["K"])
print(t,type(t[1]))
```

```python
tb2["K"].filled(np.nan)
```

```python
tb2["K"].filled(-999.)
```

```python
#vamos fazer alguns plots
import matplotlib.pyplot as plt
#scatter faz coisas diferentes do plot
#como por exemplo poder controlar o tamanho dos pontos
plt.scatter(tb2["H"]-tb2["K"],tb2["J"]-tb2["H"])
#plt.ylim(0.,0.5)
#usando LaTeX nos eixos e mudando o tamanho
plt.xlabel('$H-K$',fontsize=20)
plt.ylabel('$J-H$',fontsize=20)
plt.ylim(1.0,2.0)
plt.xlim(0.5,1.0)
plt.show()
```

```python
ra = coord.Angle(tb2['RA_J'].filled(np.nan)*u.degree)
ra = ra.wrap_at(180*u.degree)
dec = coord.Angle(tb2['DEC_J'].filled(np.nan)*u.degree)
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.scatter(ra.radian, dec.radian)
ax.grid(True)
plt.xlabel('RA (radians)',fontsize=20)
plt.ylabel('DEC (radians)',fontsize=20)
plt.show()
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.scatter(ra.deg, dec.deg)
ax.grid(True)
plt.xlabel('RA ($^h$)',fontsize=20)
plt.ylabel('DEC ($^o$)',fontsize=20)
plt.show()
```
